```python
%cd /data/c2_dataset
from glob import glob
test_img_list=glob('./test/*.png')
train_img_list=glob('./train/*.png')
valid_img_list=glob('./valid/*.png')

test_png_list=glob('./test/*.jpg')
train_png_list=glob('./train/*.jpg')
valid_png_list=glob('./valid/*.jpg')

with open('./test.txt','w') as f:
    f.write('\n'.join(test_img_list)+'\n')
    f.write('\n'.join(test_png_list)+'\n')

with open('./train.txt','w')as f:
    f.write('\n'.join(train_img_list)+'\n')
    f.write('\n'.join(train_png_list)+'\n')
    
with open('./valid.txt','w')as f:
    f.write('\n'.join(valid_img_list)+'\n')
    f.write('\n'.join(valid_png_list)+'\n')
```
